```python
#Importing the necessary modules
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Print the real image
img = cv2.imread('insta.png')
plt.figure(figsize=(8,10))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),cmap='gray')
plt.title("Real Image")
plt.show()
```

```python
# CREATING A NEW IMAGE BY CONVERTING THE REAL IMAGE TO GREYSCALE

gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,10))
plt.imshow(gray_image,cmap='gray')
plt.title("Greyscale Image")
plt.show()
```

```python
#INVERT THE IMAGE 

invert_image = 255 - gray_image
plt.figure(figsize=(8,10))
plt.imshow(invert_image,cmap='gray')
plt.title("Smoothen OR Inverted Image")
plt.show()
```

```python
#NOW WE BLUR IMAGE BY USING THE GAUSSIAN FUNCTION

blurred = cv2.GaussianBlur(invert_image, (21, 21), 0)
plt.figure(figsize=(8,10))
plt.imshow(blurred,cmap='gray')
plt.title("Blurred Image")
plt.show()
```

```python
#convert the blured image to sketch pencil image
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.figure(figsize=(5,5))
plt.imshow(pencil_sketch,cmap='gray')
plt.title("Pencil Sketch Image")
plt.show()
```
